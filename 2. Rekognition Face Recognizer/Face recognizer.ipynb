```python
#Indexing faces from the images which are stored in S3 bucket.

#Trained them with the Names with corresponding image.

bucket='facerecognizer'
collectionId='recognitionface'
s3 = boto3.client('s3', aws_access_key_id=aws_id,
                       aws_secret_access_key=aws_key)
#get all images one by one into collection
def get_s3_keys(bucket):
        keys = []
        resp = s3.list_objects_v2(Bucket=bucket)
        for obj in resp['Contents']:
            keys.append(obj['Key'])
        return keys
fileName= get_s3_keys(bucket) 
print("Training in progress...\n")

for name in fileName:
    if name.startswith('M'):
        response=client.index_faces(CollectionId=collectionId,
        Image={'S3Object':{'Bucket':bucket,'Name':name}},
        ExternalImageId='Mridushi',
        DetectionAttributes=['ALL'])
        print ('Trained...  ' + name)
    elif name.startswith('P'):
        response=client.index_faces(CollectionId=collectionId,
        Image={'S3Object':{'Bucket':bucket,'Name':name}},
        ExternalImageId='Prashant',
        DetectionAttributes=['ALL'])
        print ('Trained...  ' + name)
print("\nTraining Done ")
```

```python
#Testing an image which is stored in S3 bucket and get successfull result.
bucket='facerecognizer'
collectionId='recognitionface'
fileName='Test.jpg'
threshold = 90
maxFaces=1

client=boto3.client('rekognition')


response=client.search_faces_by_image(CollectionId=collectionId,
                        Image={'S3Object':{'Bucket':bucket,'Name':fileName}},
                        FaceMatchThreshold=threshold,
                        MaxFaces=maxFaces)

                            
faceMatches=response['FaceMatches']
print ('Matching faces')
for match in faceMatches:
    print ('FaceId:' + match['Face']['ExternalImageId'])
    print ('Confidence: ' + "{:.2f}".format(match['Similarity']) + "%")
    print ()
```

```python
#Now capture image via Webcam and call search_face_by_image method which use Index to identify face in the Collection bucket.

#If the face did not recoginze then it will response with a statement.

import cv2
import sys
import boto3
import time
import os
from PIL import Image
from IPython.core.display import display

cam=cv2.VideoCapture(0)
time.sleep(0.1)
check, frame = cam.read()
try:
    cv2.imwrite("compare.jpg",frame)
    display(Image.open("compare.jpg"))
except OSError:
    print("Restart Kernel")
    
if __name__ == "__main__":
    try:
        bucket='facerecognizer'
        collectionId='recognitionface'
        maxFaces=1
        client=boto3.client('rekognition')
        response=client.search_faces_by_image(CollectionId=collectionId,
                    Image={'Bytes': cv2.imencode('.jpg', frame)[1].tobytes()},
                    FaceMatchThreshold=90, MaxFaces=maxFaces)
    except cv2.error as e:
        print("Can\'t do the thing")
        sys.exit(1)
    
    faceMatches=response['FaceMatches']
    for match in faceMatches:
        print ('This person is ' + match['Face']['ExternalImageId'])
        print ('Confidence: ' + "{:.2f}".format(match['Similarity']) + "%")
    if faceMatches==[]:
        print("Unable to recognize the person")
    
        
    
del(cam)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
