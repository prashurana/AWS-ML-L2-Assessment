```python
import cv2
import boto3
from IPython.core.display import display
from PIL import Image

aws_id='AKIAIX72CBE63C5U5NQA'
aws_key='Tr6qxpcW/zXXVVt7w0S4lpZ552h7iBjm9GARYlh0'

img="./test.jpg"

try:
    frame = cv2.imread(img)
    image_bytes = cv2.imencode('.jpg', frame)[1].tobytes()
except IOError as e:
    print ("I/O error({0}): {1}".format(e.errno, e.strerror))
    exit(-1)



if __name__ == "__main__":
    client=boto3.client('rekognition', region_name='us-east-1',
                       aws_access_key_id=aws_id,
                       aws_secret_access_key=aws_key)
    
    response = client.detect_labels(Image={'Bytes': image_bytes}, MinConfidence=50)
    
    label = "Football"  # Here, I have chosen Glasses as Specific Object.
    
    print('Detecting Football\n')
    
    def search(label, resp):
        return [obj for obj in resp if obj["Name"] == label]
    
    resp = search(label, response['Labels'])
    
    if resp == []:
        print("{} is missing in this image.".format(label))
    else:
        print("Yay! I have found the {}.".format(label))
        
    display(Image.open(img))

```
